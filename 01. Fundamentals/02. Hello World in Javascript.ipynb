## `console.log()`

- This is a print statement in Javascript.
- It logs data in browser console.

```javascript
console.log("Hello World")
```

```javascript
console.log("Hi, I am Parth Maru.")
console.log("I am very much excited to learn Javascript!!!")
```

### Steps to use `console.log()` in browser

- Open any browser, chrome/ firefox most preferrable.
- Open the developer console.
- Go to console section.
- Type `console.log("Hello World")` and click enter.
